**How many people have the word Chief in their job title ?**

Note: This is pretty tricky

Answer: 477

````python
num_chiefs = sum('chief' in title.lower().split() for title in sal['JobTitle'])
num_chiefs
````

````python
# consider case sensitive, must be only Chief
chief_titles = sal[sal['JobTitle'].str.contains('Chief', na=False)]

# Drop duplicates based only on EmployeeName
unique_chief_titles = chief_titles.drop_duplicates(subset='EmployeeName')

# Count the number of unique entries
chief_titles_count = unique_chief_titles.shape[0]
chief_titles_count
````

````python
# consider case insensitive
chief_titles = sal[sal['JobTitle'].str.contains('Chief', case=False, na=False)]

# Drop duplicates based only on EmployeeName
unique_chief_titles = chief_titles.drop_duplicates(subset='EmployeeName')

# Count the number of unique entries
chief_titles_count = unique_chief_titles.shape[0]
chief_titles_count
````

**Hard: How many people have a credit card that expires in 2025 ?**

Answer: 1033

````python
# First, ensure 'CC Exp Date' is a string
ecom['CC Exp Date'] = ecom['CC Exp Date'].astype(str)

# Convert 'MM/YY' to 'YYYY-MM-01', assuming we are dealing with the last year of the card's validity
ecom['CC Exp Date'] = ecom['CC Exp Date'].apply(lambda x: f'20{x[-2:]}-{x[:2]}-01')

# Convert to datetime format
ecom['CC Exp Date'] = pd.to_datetime(ecom['CC Exp Date'], format='%Y-%m-%d', errors='coerce')

# Filter for the year 2025
expiring_2025 = ecom[ecom['CC Exp Date'].dt.year == 2025]

# Count the number of rows
num_people = expiring_2025.shape[0]
print(num_people)
````
