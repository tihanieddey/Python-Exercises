**Problem 3**

Get number of items as input and generate that many ADAM Numbers

```python
numbers = int(input("Enter a number you want to generate Adam Numbers: "))

count = 0
num = 1

while count < numbers:
    reverse_num = 0
    temp = num
    while temp > 0:
        reverse_num = (reverse_num * 10) + (temp % 10)
        temp //= 10

    square_num = num ** 2
    square_reverse = reverse_num ** 2

    rev_sqrevnum = 0
    temp = square_reverse
    while temp > 0:
        rev_sqrevnum = (rev_sqrevnum * 10) + (temp % 10)
        temp //= 10
    if square_num == rev_sqrevnum:
        print(num)
        count += 1
    num += 1
```

**Problem 4**

Get number of items as input and generate that many Armstrong Numbers

```python
numbers = int(input("Enter a number you want to generate Armstrong Numbers:"))

count = 0
num = 1

while count < numbers:
  temp = num
  total_digits = 0
  total = 0

  while temp > 0:
    total_digits += 1
    temp //= 10

  temp_sum = num
  while temp_sum > 0:
    remain = temp_sum % 10
    total += remain ** total_digits
    temp_sum //= 10

  if total == num:
    print(num)
    count += 1
  num += 1
```

**Problem 10**

Write a Python program that takes a string as input, which contains numbers separated by commas. Convert the string to a list of numbers and determine whether all the numbers are different from each other.

```python
n = [int(x) for x in\
     input("Enter a list of numbers separated by commas: ").split(',')]

count = 0
for i in n:
  count += 1

has_duplicate = False

for i in range(count):
  for j in range(i +1, count):
    if n[i] == n[j]:
      has_duplicate = True
      break
  if has_duplicate:
    break

if has_duplicate:
  print("There are duplicates in the list of numbers given")
else:
  print("The numbers are different from each other")
```

**Problem 14**

Write a python program which takes a number as input and convert the number to binary. Note: Don't use any builtin functions.

```python
number = input("Please enter a number you want to change into binary: ")

temp = int(number)
result = ""

if temp == 0:
  print("Binary for 0 is 0")
else:
  while temp > 0:
    result = str(temp % 2) + result
    temp //= 2

print(result)
```

**Problem 15**

Write a python program which takes a binary number as input and convert the number to decimal. Note: Don't use any builtin functions.

```python
number = input("Please enter a number you want to change into decimal: ")

temp = int(number)
result = 0
n = 0

if temp == 0:
  print("Decimal for 0 is 0")
else:
  while temp > 0:
    result = result + ((2 ** n) * (temp % 10))
    temp //= 10
    n += 1

print(result)
```
